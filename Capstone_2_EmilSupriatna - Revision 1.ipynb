- ***Handling Missing Values***

```python
# Menghitung berapa persen missing value pada kolom tertentu

print(round(df.isna().mean()*100,2))
```

```python
# Menghitung total varian data unik,
# selanjutnya diputuskan missing handling dengan modus/median/mean

columns = ['store_and_fwd_flag', 'RatecodeID', 'passenger_count',
           'payment_type', 'trip_type', 'congestion_surcharge']

for i in columns:
    print(f"unique value for {i}:  {df[i].nunique()}")
```

```python
# Karena varian data unik berjumlah sedikit, handling dilakukan dengan modus
# Memeriksa persentase data unik non missing untuk mengetahui modus data

columns = ['store_and_fwd_flag', 'RatecodeID', 'passenger_count',
           'payment_type', 'trip_type', 'congestion_surcharge']

for column in columns:
    print(f"Value counts for {column}:")
    print(round(df[column].value_counts().sort_index()/df.shape[0]*100,2))
    print('\n')
```

- ***Handling Outliers***

Pada analisis deskriptif yang telah disampaikan, terdapat nilai *outliers* yang sangat ekstrim di sejumlah kolom. *Outliers* sendiri merupakan data pencilan yang terletak jauh dari pusat data saat dilihat dengan diagram tertentu. Atau dengan kata lain, *outliers* adalah nilai-nilai yang secara signifikan berbeda dari sebagian besar data dalam satu set. Misalnya dalam suatu perusahaan, terdapat gaji karyawan sebesar 5juta, 5.5juta, 7juta dan 20juta untuk gaji CEO. Maka, gaji 20juta dapat dikatakan sebagai *outliers*.

*Outliers* dapat muncul karena kesalahan pengukuran ataupun memang data yang sebenarnya seperti data gaji di atas. Identifikasi *outliers* dapat dilakukan dengan metode statistik seperti IQR (Interquartile Range) dan z-score atau visualisasi seperti box plot dan histogram. Keberadaan outliers bisa memiliki dampak signifikan pada hasil analisis statistik, seperti mean (rata-rata) dan standar deviasi. Untuk mencegah bias pada analisis, *outliers* dapat ditangani salah satunya dengan cara penghapusan data. Meski demikian, penghapusan *outliers* harus dilakukan dengan hati-hati, dan dengan catatan bahwa alasan keberadaan outliers perlu dipahami dengan baik terlebih dahulu.

```python
# Filter data yang memiliki variasi data unik di atas 100
# Karena data dengan variasi data unik sedikit cenderung rentan dianggap sebagai outliers

for i in df.columns:
    if df[i].nunique()>100:
        print(f"unique value for {i}:\t{df[i].nunique()}")
```

```python
# Dengan mengecualikan data tanggal (datetime) dan LocationID,
# identifikasi outliers dengan visualisasi distplot dari seaborn

columns = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount']

plt.figure(figsize=(12, 6))

for i, column in enumerate(columns, start=1):
    plt.subplot(2, 2, i)
    sns.distplot(df[column])

plt.tight_layout()
plt.show()
```

Pada grafik, terlihat outliers yang sangat ekstrim. Misal pada kolom trip_distance, data terpusat mendekati angka 0. Sedangkan, terdapat sebaran data hingga sebesar 120 ribu mil. Atau pada kolom fare_amount dan total_amount. Rentang data berkisar dari 0 hingga 100. Namun, terdapat data tersebar hingga 500 USD. Sebaran data ekstrim seperti ini dapat dikategorikan tidak normal dan pada kasus ini dapat diasumsikan sebagai kesalahan dalam proses input. Oleh karena itu, keputusan menghapus data outlier dapat dilakukan untuk menghindari bias dalam proses analisis. Sebab, nilai-nilai *outlier* ini dapat mengganggu nilai rata-rata dan standard deviasi data.

```python
# Menghitung jumlah dan persentase outliers dengan metode IQR

columns_outliers = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount']

for column in columns_outliers:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    print(f'Outliers pada {column} adalah {outliers.count()} atau sebesar {round(outliers.count()/df.shape[0]*100,2)}%')
```

```python
# Menghapus Outliers

for column in columns_outliers:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df.reset_index(drop=True, inplace=True)
```

```python
# Tampilkan kembali visualisasi distplot, maka data terlihat mendekati distribusi normal
# Namun, untuk mengatakan data terdistribusi normal atau tidak, perlu dilakukan uji normalitas

plt.figure(figsize=(12, 6))

for i, column in enumerate(columns_outliers, start=1):
    plt.subplot(2, 2, i)
    sns.distplot(df[column])

plt.tight_layout()
plt.show()
```

```python
# Bandingkan dengan menggunakan boxplot
# Meskipun masih terdapat data outlier di luar 'upper bound', namun sebarannya tidak terlalu ekstrim

plt.figure(figsize=(12, 6))

for i, column in enumerate(columns_outliers, start=1):
    plt.subplot(2, 2, i)
    sns.boxplot(x = df[column])

plt.tight_layout()
plt.show()
```

```python
df.shape
```

```python
# Melakukan uji normalitas pada ketiga kolom yang tampak terdistribusi secara normal
# metode Kolmogorov-Smirnov merupakan metode umum yang digunakan untuk uji normalitas dan dengan ukuran data besar

norm_test = ['trip_distance', 'fare_amount', 'total_amount']

for column in norm_test:
    statistic, p_value = kstest(df[column], 'norm')

    # Interpretasi p-value
    alpha = 0.05
    if p_value > alpha:
        print(f"Data '{column}' terdistribusi normal (tidak dapat menolak H0)\n")
    else:
        print(f"Data '{column}' tidak terdistribusi normal (menolak H0)\n")
```

Nt. Ketiga variabel yang diujikan tidak terdistribusi secara normal. Dengan demikian dapat dikatakan bahwa distribusi data akan mengacu pada grafik sebelumnya, yakni terdistribusi secara *Right Skew*

#### 3) *Rekayasa Fitur/Kolom*

Dalam konteks Exploratory Data Analysis (EDA), rekayasa fitur/kolom dapat membantu untuk memahami data lebih dalam dan meningkatkan keefektifan analisis eksploratif. Salah satu bentuk rekayasa fitur adalah ekstraksi fitur waktu, yakni memisahkan tanggal menjadi beberapa kategori waktu seperti tahun, bulan dan hari.

Ekstraksi fitur waktu pada projek ini juga penting dilakukan dalam rangka mendapatkan pemahaman *"Customer Behavior*" yang lebih baik berdasarkan aspek perilaku harian dan juga perilaku dalam satuan waktu 24 jam. Kolom yang akan ditambahkan antara lain kolom tanggal *pickup*, jam, nama hari, kategori hari (weekend/weekdays) dan durasi waktu.

```python
# Menambah beberapa kolom baru
df['pickup_date'] = pd.to_datetime(df['lpep_pickup_datetime']).dt.date
df['days'] = df['lpep_pickup_datetime'].dt.day_name()
df['day_number'] = df['lpep_pickup_datetime'].dt.dayofweek + 1
df['days_category'] = df['lpep_pickup_datetime'].dt.day_name()\
    .apply(lambda x: 'weekend' if x in ['Saturday', 'Sunday'] else 'weekdays')
df['pickup_hour'] = df['lpep_pickup_datetime'].dt.hour
df['time_duration'] = df['lpep_dropoff_datetime'] - df['lpep_pickup_datetime']
df['time_duration'] = df['time_duration'].dt.total_seconds() / 60
```

Kolom baru bernama "time_duration" bertipe numerik. Oleh sebab itu, pengecekan distribusi data menjadi penting untuk kebutuhan analisis tingkat lanjut. Ketika  ditemukan data ekstrim, kita perlu handling. Sehingga, data yang tampil hanya data harian normal saja. 

```python
plt.figure(figsize=(8, 4))
sns.distplot(df['time_duration'])

plt.tight_layout()
plt.show()
```

```python
# Mendeteksi dan menghitung total outlier pada kolom "time_duration"

Q1 = df['time_duration'].quantile(0.25)
Q3 = df['time_duration'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = df[(df['time_duration'] < lower_bound) | (df['time_duration'] > upper_bound)]['time_duration']
print(f'time_duration = {outliers.count()} - {round(outliers.count()/df.shape[0]*100,2)}%')
```

```python
# Terdapat 2.65% ourlier. Hapus data, kemudian periksa kembali dalam bentuk histogram.

df = df[(df['time_duration'] >= lower_bound) & (df['time_duration'] <= upper_bound)]

df.reset_index(drop=True, inplace=True)

sns.distplot(df['time_duration'])
plt.show()
```

```python
# Seperti pada kolom sebelumnya, lakukan uji normalitas pada kolom "time_duration"

statistic, p_value = kstest(df['time_duration'], 'norm')

# Interpretasi p-value
alpha = 0.05
if p_value > alpha:
    print(f"Data 'time_duration' terdistribusi normal (tidak dapat menolak H0)\n")
else:
    print(f"Data 'time_duration' tidak terdistribusi normal (menolak H0)\n")
```

Berdasarkan pengujian data, tidak ada satu pun yang terdistribusi secara normal, melainkan berbentuk *right skew* seperti yang tampak pada visualisasi yang telah dilakukan.

```python
def viz_01():
    grid = plt.GridSpec(2, 2, wspace=0.2, hspace=0.4)
    plt.figure(figsize=(12, 6))

    # heatmap: days vs times
    plt.subplot(grid[0:, 1])
    df_heatmap = pd.pivot_table(df, values='total_amount',
                                index='pickup_hour',
                                columns=['day_number','days'],
                                aggfunc='mean')
    sns.heatmap(df_heatmap, cmap="YlGnBu", annot=True, fmt=".2f", cbar_kws={'label': 'Value'})
    plt.title('Average Daily Amount in 24 hours', fontsize=11)
    plt.yticks(rotation=0)

    # Lineplot: weekly vs weekend
    plt.subplot(grid[0, 0])
    df_line = df.groupby(['days_category', 'days', 'pickup_hour'])['passenger_count'].count().reset_index()
    df_line = df_line.groupby(['days_category', 'pickup_hour'])['passenger_count'].mean().reset_index()
    sns.lineplot(data=df_line, x='pickup_hour', y='passenger_count', hue='days_category')
    plt.legend(loc='upper right', fontsize=6)
    plt.axvspan(13, 20, alpha=0.2, color='green')
    plt.title('Weekdays vs Weekend Average Daily Passenger', fontsize=11)
    plt.ylabel('Count of Passengers')
    plt.xticks(df_line['pickup_hour'])
    plt.grid(which='both', axis='both', linestyle='--', linewidth=0.5)
    plt.xticks(rotation=90)

    # Bar plot: daily average
    plt.subplot(grid[1, 0])
    df_bar = df.groupby(['day_number','pickup_hour'])[['time_duration']].mean().reset_index()
    sns.barplot(data=df_bar, x='day_number', y='time_duration')
    plt.ylabel('Average Time Duration')
    plt.xlabel('Number of Day, Start from Monday: 1')
    plt.title('Daily Mean Time Duration', fontsize=11)

    plt.savefig('viz_01.png')
    plt.show()
```

```python
def viz_02():
    grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.6)
    plt.figure(figsize=(7, 7))

    # Heatmap: 3 Variable Correlation
    plt.subplot(grid[0, 0])
    corr = df[['total_amount','time_duration', 'trip_distance']].corr(method='spearman')
    heatmap = sns.heatmap(corr, annot=True)
    plt.title('Variable Correlation', fontsize=11)
    heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=9, rotation=45)
    heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=9)

    # Pieplot: Interval Time Duration Proportion
    plt.subplot(grid[0, 1])
    df_interval = df[['days_category','days','pickup_hour', 'time_duration',  'PULocationID']]
    df_interval['time_interval'] = pd.cut(df_interval['time_duration'],
                                          bins=[0, 9, 18, 30],
                                          labels=['0-9 menit', '10-18 menit', '>18 menit'],
                                          right=False)
    df_pie = df_interval.groupby('time_interval')[['time_duration']].count().reset_index()
    sns.color_palette("pastel")
    plt.pie(df_pie['time_duration'], labels=df_pie['time_interval'], autopct='%1.1f%%')
    plt.title('Duration Interval Proportion', fontsize=11)

    # Barplot: Top 20 PULocationID during January 2023
    plt.subplot(grid[1, 0:])
    Top10_PULocationID = df_interval[df_interval['time_interval'] == '10-18 menit']['PULocationID'].value_counts().head(20)
    Top10_PULocationID = pd.DataFrame(Top10_PULocationID).reset_index().rename(columns={'index':'location'})
    sns.barplot(x=Top10_PULocationID['location'], y=Top10_PULocationID['PULocationID'])
    plt.text(0, Top10_PULocationID['PULocationID'][0], str(Top10_PULocationID['PULocationID'][0]), ha='center', va='top', rotation=90)
    for i, value in enumerate(Top10_PULocationID['PULocationID'][1:], start=1):
        plt.text(i, value, str(value), ha='center', va='bottom', rotation=90)
    plt.ylabel('Count of Passengers')
    plt.xlabel('Pick Up Location ID')
    plt.title('Top 20 PULocationID during January 2023', fontsize=13)
    plt.xticks(rotation=90)


    plt.savefig('viz_02.png')
    plt.show()
```

#### Visualization 3: Credit Card, Peluang Besar Memperoleh Tip

```python
def viz_03():
    grid = plt.GridSpec(1, 2, wspace=0.1, hspace=0.4)
    plt.figure(figsize=(12, 4))

    # Heatmap: Correlation of Tip Amount with related variable
    plt.subplot(grid[0, 0])
    corr = df[['tip_amount', 'time_duration', 'trip_distance', 'payment_type', 'total_amount', 'congestion_surcharge']].corr(method='spearman')
    total_amount_corr = corr.loc[['tip_amount']]
    sns.heatmap(total_amount_corr.T, annot=True)
    plt.title('Tip Amount corr. with related variable', fontsize=11)

    # Table: Explorasi Payment Type
    plt.subplot(grid[0, 1])
    table = df[['payment_type','total_amount', 'tip_amount']]
    table['customer'] = 1
    table['giving_tip'] = df['tip_amount'].apply(lambda x: 1 if x > 0 else 0)
    table[['giving_tip']].value_counts().sort_index()
    table = table[['payment_type', 'customer', 'giving_tip']]
    table_2 = table.groupby('payment_type')[['customer','giving_tip']].sum().reset_index()
    table_2['perc.(%)'] = round(table_2['giving_tip']/table_2['customer']*100,2)
    table_2[['customer', 'giving_tip']] = table_2[['customer', 'giving_tip']].applymap('{:,.0f}'.format)
    mapping = {1: 'Credit card', 2: 'Cash', 3: 'No charge', 4: 'Dispute', 5: 'Unknown'}
    table_2['payment_type'] =table_2['payment_type'].replace(mapping)
    table = plt.table(cellText=table_2.values, colLabels=table_2.columns, loc='center', cellLoc='center', bbox=[0, 0, 1, 1])
    header_colors = ['#ff7f50'] * len(table_2.columns)
    for i, key in enumerate(table_2.columns):
        cell = table[(0, i)]
        cell.set_facecolor(header_colors[i])
        cell.get_text().set_weight('bold')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 3)
    plt.axis('off')
    plt.title('Tip Amount Based on Payment Type', fontsize=11)

    plt.savefig('viz_03.png')
    plt.show()
```

Meskipun korelasi tidak menyiratkan kausalitas (sebab-akibat), analisis ini dapat memberikan petunjuk tentang hubungan antara dua variabel.
